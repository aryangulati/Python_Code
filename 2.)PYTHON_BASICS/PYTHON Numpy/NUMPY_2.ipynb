## Vectorization

```python
a=np.array([2,3])
b=np.array([5,4])
```

```python
b[0]
```

```python
def distance(a,b):
    dx=b[0]-a[0]
    dy=b[1]-a[1]
    
    return np.sqrt(dx**2+dy**2)
```

```python
distance(a,b)
```
