#### [matplotlib.pyplot.xscale](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xscale.html)

You are setting the scale of your y and x-axis to be logarithmic in scale.

Normally every y distance on your axis your values increment by a fixed amount, example:

   0 at 0cm
   1 at 1cm
   2 at 2cm  
  ...
1000 at 10m
With logarythmic the values scale by a magnitude. Example for powers of 10:

   0 at 0cm
   1 at 1cm
  10 at 2cm
 100 at 3cm
1000 at 4cm etc.
It is a way to display widely spread data in a compacter format.

See logarithmic scale on wikipedia

Your data has a cluster of values and an outlier - by printing with a logarithmic scale your blob gets shown over distance whatever and the big distance between the blob and the outlier takes less screenarea due to it being logarithmic.

```python
plt.scatter(data1_x, data1_y, label= "small range", marker="^")
plt.scatter(data2_x, data2_y, label= "big range")
plt.xscale('log')
plt.legend()
plt.show()
```

# Bar plot

```python
x=["car1","car2","car3"]
sales2020=[20,5,13]
sale2019=[29,6,18]
```

```python
plt.bar(x,sales2020,width=-0.3,align="edge")
plt.bar(x,sale2019,width=0.2)
plt.show()
```

```python
x = np.array([1,2,3]) #if we make this a numpy arry then we can shift 
sales_2019 = [20,5,13]
sales_2020 = [12, 10, 21]
```

```python
plt.bar(x, sales_2019, width=0.2, align='edge')
plt.bar(x+0.2, sales_2020, width=0.2, align='edge')
plt.show()
```
