```python
x_cord=np.array([0,1,2])
plt.bar(x_cord,[10,20,15]) 
plt.bar(x_cord+0.5,[20,15,10]) 
#here we can see that bar overlap each other 
plt.show()
```

```python
#so for that if we multiply x axis by 2 
# and also give parameter of width to bar.plot()
x_cord=np.array([0,1,2])*2
plt.bar(x_cord,[10,20,15],width=0.5,label="Current Year",tick_label=["BMW","Mustang","AUDI"]) 
plt.bar(x_cord+0.5,[20,15,10],width=0.5,label="Next Year") 
plt.title("Car price Comparision")
plt.xlabel("Price")
plt.ylabel("CARS")
plt.legend()
plt.show()
```

```python
#so for that if we multiply x axis by 2 
# and also give parameter of width to bar.plot()
plt.style.use("dark_background")
x_cord=np.array([0,1,2])*2
plt.bar(x_cord,[10,20,15],width=0.5,label="Current Year",color="orange",tick_label=["BMW","Mustang","AUDI"]) 
plt.bar(x_cord+0.5,[20,15,10],width=0.5,label="Next Year") 
plt.ylim(0,40)
plt.xlim(0,6)
plt.legend()
plt.show()
```

```python
xsn = np.random.randn(100)
sigma=5
u=60
X=xsn*sigma+u
print(X) #most values are in range 55 to 65
```

```python
plt.hist(X)
plt.ylabel("Prob/Freq of Buyied cars")
plt.xlabel("Cars Range")
plt.show()
```

# Movies Datasets
- plots the length of Movie Title Names on X-axis
- Frequency Counts on Y axis is Number of Movies having "x" characters in their title

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("movie.csv")
```

```python
df.head(n=11)
```

```python
df.columns
```

```python
df.shape
```

```python
titles=df.get("movie_title")
```

```python
print(titles[53])
```

```python
titles=list(df.get("movie_title"))
```

```python
type(titles)
```

```python
print(titles[:5])
```

```python
freq_titles={}
for title in titles:
    length =len(title)
    
    if freq_titles.get(length) is None:
        freq_titles[length]=1
    else:
        freq_titles[length]+=1
        
    
```

```python
print(freq_titles)
```

```python
x=np.array(list(freq_titles.keys()))
y=np.array(list(freq_titles.values()))
print(x,y)
```

```python
plt.scatter(x,y)
plt.xlabel("Length of movie Title")
plt.ylabel("Freq Count")
plt.title("Movie Data ")
```
