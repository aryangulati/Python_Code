# jointplot

```python
#if you want to compare 2 varaible or feature
#both variable on rows and column 


sns.jointplot(x="total_bill",y="tip" ,data =tips)
#its used to draw some relation between two things
#also this trend is positive
#both the distribution we get from dist graph we get it here !!
```

```python
# there is a linear trend in total bill and tip
# as bill increase tip also  increase
```

```python
sns.jointplot(x="total_bill",y="tip" ,data =tips,kind="scatter")
sns.jointplot(x="total_bill",y="tip" ,data =tips,kind="hex") #more density
sns.jointplot(x="total_bill",y="tip" ,data =tips,kind="kde")
sns.jointplot(x="total_bill",y="tip" ,data =tips,kind="reg")

```

```python
sns.pairplot(tips)
#it compare all the features together
#pair wise co rlation with each other
```

```python
#if you want data differently for male and female you can plot it with Hue 
sns.pairplot(tips,hue="sex")
```
