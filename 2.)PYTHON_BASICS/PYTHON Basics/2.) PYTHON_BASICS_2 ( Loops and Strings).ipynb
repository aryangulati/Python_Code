S.format_map(mapping) -> str

Return a formatted version of S, using substitutions from mapping.
The substitutions are identified by braces ('{' and '}').
Type:      builtin_function_or_method

The format_map(mapping) is similar to str.format(**mapping) method.**

The only difference is that str.format(**mapping) copies the dict whereas str.format_map(mapping) makes a new dictionary during method call. This can be useful if you are working with a dict subclass.**

source:https://www.programiz.com/python-programming/methods/string/format_map

```python
class Coordinate(dict):
    def __missing__(self, key):
      return key


print('({x}, {y})'.format_map(Coordinate(x='6')))
print('({x}, {y})'.format_map(Coordinate(y='5')))
print('({x}, {y})'.format_map(Coordinate(x='6', y='5')))
```

## Decoraters 

```python
users={
    "aryan":"password",
    "ram":"pass",
    "krishna":"hello"
}
```

```python
def show(username,password):
    if username in users and users[username] == password:
        print("hello World")
    else:
        print("nOt authenticated")
```

```python
show("aryan","passs")
```

```python
show("ram","pass")
```

```python
def add(username,password,a,b):
    if username in users and users[username] == password:
        print(a+b)
    else:
        print("Not authenticated")
```

```python
add("aryan","password",1,2)
#so we can see their is reductancy in function
```

```python
#so function are also an obj
def login_required(func):
    def wrapper(username,password,*args,**kwargs):
        if username in users and users[username] == password:
            print(func(*args,**kwargs))
        else:
            print("Not authenticated")
    return wrapper       

        
    
```

```python
add= lambda a,b:a+b
add=login_required(add)
#both of the blocks above and below are same 
#they have same functioning
#creates the function add 
#then pass it the function login required
#change sthe variable of add function 
#modified the var with the new func return
```

```python
@login_required

def add(a,b):
    print(a+b)
```

```python
add(1,2)
```

```python
protected_add= login_required(add)
#generic function and we are just passing a function as parameter
```

```python
print(protected_add) #so now are function is password protected
```

```python
protected_add("ram","pass")
```

```python
protected_add("ram","pass",2,3)
```
