```python
import cv2 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
eyes_cascade=cv2.CascadeClassifier("Train/third-party/frontalEyes35x16.xml")
```

```python
nose_cascade=cv2.CascadeClassifier("Train/third-party/Nose18x15.xml") 
```

```python
img = cv2.imread("Train/Jamie_Before.jpg")
```

```python
while True:
        face= cv2.imread("Train/Jamie_Before.jpg",)
        specs=cv2.imread("Train/glasses.png",)
        mustache=cv2.imread("Train/mustache.png")

        eyes = eyes_cascade.detectMultiScale(img, 1.5, 5)
        nose = nose_cascade.detectMultiScale(img, 1.9, 5)

        for eye in eyes:
            x, y, w, h = eye

            cv2.rectangle(img, (x,y), (x+w, y+h), (255, 0, 0), 2)
        
        for n in nose:
            x, y, w, h = n

            cv2.rectangle(img, (x,y), (x+w, y+h), (255, 0, 0), 2)
            
            
            
            
        cv2.imshow("window", img)
        key = cv2.waitKey(1) #1ms wait by waitkey function
        if key == ord("q"):
            break
cv2.destroyAllWindows()
```
